```python
import pandas as pd
```

```python
df_01_09 = pd.read_excel('/content/reestr_01_09.xlsx', sheet_name='list1', names=['surname', 'name1', 'father_name1', 'date1', 'id1'], header=None)
df_01_09['fullname'] = [(str(df_01_09.surname[i]).strip() + str(df_01_09.name1[i]).strip()).lower() for i in range(len(df_01_09))]
df_01_09.head()
```

```python
df_last = pd.read_excel('/content/result_last_reestr.xlsx', sheet_name='list1', names=['surname', 'name2', 'father_name2', 'date2', 'id2'], header=None)
df_last['fullname'] = [(str(df_last.surname[i]).strip() + str(df_last.name2[i]).strip()).lower() for i in range(len(df_last))]
df_last.head()
```

```python
df1 = pd.DataFrame(df_01_09.fullname).merge(pd.DataFrame(df_last.fullname), how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']
df1 = df1.merge(df_01_09, how='left', on='fullname').drop_duplicates()
df1.head()
```

```python
df2 = pd.DataFrame(df_01_09.fullname).merge(pd.DataFrame(df_last.fullname), how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='right_only']
df2 = df2.merge(df_last, how='left', on='fullname').drop_duplicates()
df2.head()
```

```python
df1.to_excel('/content/df1.xlsx')
```

```python
df2.to_excel('/content/df2.xlsx')
```
