```python
import pandas as pd
from scipy.stats import spearmanr
from numpy import nan
```

```python
df = pd.read_excel("/content/good_data.xlsx")
```

```python
df.head(15)
```

```python
def get_non_na_series(df, values, s_1, s_2):
  my_index = 0
  i = 0
  for val in values:
    if not val == nan:
      s_2.append(val)
      s_1.append(df["36_ПКВ"][i])
      my_index += 1
    i += 1
```

```python
for name, values in df.iteritems():
  s_1 = pd.Series()
  s_2 = pd.Series()
  get_non_na_series(df, values, s_1, s_2)
  coef, p = spearmanr(s_1, s_2)
  print(name, coef, p)
```
